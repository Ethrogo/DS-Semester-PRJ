1. What is in your data?
The dataset includes game-level statistics and team/player performance metrics collected from multiple sources. Key features include gameid, team, home, score, rebounds, and assists, as well as advanced metrics like offensive rating, defensive rating, and true shooting percentage. The data spans multiple seasons, covering over 2,000 games and over 30,000 individual player entries, with various metrics that capture both team and individual performance on offense and defense

```python
import pandas as pd 
import os 

csv_files = [file for file in os.listdir("data/") if file.endswith('.csv')]
dataframes = []

for file in csv_files: 
    df = pd.read_csv(os.path.join("data/", file))
    print(df.columns)  # List column names
    print(df.dtypes)   # Show data types
    print(df.describe())  # Summary statistics (min, max, mean, etc.)
    print(df.isnull().sum())  # Count missing values per column
    dataframes.append(df)
```

```python
import matplotlib.pyplot as plt

# Sample histogram
dataframes[0]['OFFRTG'].hist(bins=20)
plt.title("Distribution of OFFRTG")
plt.xlabel("OFFRTG")
plt.ylabel("Frequency")
plt.show()

# Sample bar plot for a categorical variable
dataframes[0]['team'].value_counts().plot(kind='bar')
plt.title("Distribution of Teams")
plt.xlabel("Team")
plt.ylabel("Frequency")
plt.show()
```
